## Rule for Matrix multiplication
1. The inner dimensions must match:
* ( 3,2) @ (3,2) won't work (@ is another function for matrix multiplication like torch.matmul, e.g tensor 2 tensor)

* (3,2) @ (2,3) will work
* (2,3) @ (3,2) will work

2. THe resulting matrix has the shape of the **outer dimenstions**:

* (2,3) @ (3,2) -> (2,3)
* (3,2) @ (2,3) -> (3,3)


````python
torch.matmul(torch.rand(3,2), torch.rand(2,3))
````

### Shape Error

````python
#shapes for matrix mulriplication
tensor_A = torch.tensor([[1,2],
                         [3,4],
                         [5,6]])
tensor_B = torch.tensor([[7,10],
                         [8,11],
                         [9,12]])
#torch.mm is an alias for torch.matmul
torch.mm(tensor_A, tensor_B)
````
