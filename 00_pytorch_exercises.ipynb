```python
torch.cuda.manual_seed(1234)
```

```python
#creating 2 tensors and sending them to GPU

torch.manual_seed(1234)
c = torch.rand(2,3)

torch.manual_seed(1234)
d = torch.rand(2,3)

c_cuda = c.to("cuda")
d_cuda = d.to("cuda")
```

```python
#matrix multiplication

e = c_cuda @ d_cuda.T
```

```python
#min and max
print(f"The smallest value in e (the matrix multiplication) is {e.min()}")
print(f"The largest value in e (the matrix multiplication) is {e.max()}")
```
